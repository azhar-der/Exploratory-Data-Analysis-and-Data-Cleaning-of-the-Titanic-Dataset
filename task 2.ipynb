```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the datasets
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
gender_submission_df = pd.read_csv('gender_submission.csv')
```

```python

# Fill missing values in the 'Age' column with the median age
train_df['Age'].fillna(train_df['Age'].median(), inplace=True)

# Fill missing values in the 'Embarked' column with the most frequent value (mode)
train_df['Embarked'].fillna(train_df['Embarked'].mode()[0], inplace=True)

# Drop the 'Cabin' column since it has too many missing values
train_df.drop(columns=['Cabin'], inplace=True)
```

```python
# Exploratory Data Analysis (EDA)
# Visualize the distribution of survival
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', data=train_df)
plt.title('Distribution of Survival')
plt.show()
```

```python
# Visualize the distribution of passengers by class
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', data=train_df)
plt.title('Distribution of Passengers by Class')
plt.show()
```

```python
# Visualize the relationship between age and survival
plt.figure(figsize=(10, 8))
sns.histplot(data=train_df, x='Age', hue='Survived', multiple='stack', kde=True)
plt.title('Age Distribution by Survival Status')
plt.show()
```

```python
# Visualize the relationship between sex and survival
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=train_df)
plt.title('Survival Rate by Gender')
```

```python
# Visualize the relationship between fare and survival
plt.figure(figsize=(10, 8))
sns.histplot(data=train_df, x='Fare', hue='Survived', multiple='stack', kde=True)
plt.title('Fare Distribution by Survival Status')
plt.show()
```

```python

# Visualize correlation matrix for only numeric columns
plt.figure(figsize=(12, 10))
```

```python

# Select only numeric columns
numeric_cols = train_df.select_dtypes(include=[np.number])

# Compute the correlation matrix
corr_matrix = numeric_cols.corr()

# Plot the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Summary statistics
print("\nSummary Statistics:")
print(train_df.describe())

# Saving the cleaned dataset
train_df.to_csv('cleaned_train.csv', index=False)
```
